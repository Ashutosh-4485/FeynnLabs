# 2nd Dataset

```python
# Load your dataset
data = pd.read_excel('ev_sales.xlsx')

# Extract relevant columns for usage pattern
X_usage = data[['2 W', '3 W', '4 W', 'TOTAL']]

# Convert 'YEAR' to timestamp
X_usage['YEAR'] = pd.to_datetime(data['YEAR']).astype('int64') // 10**9

# Standardize the data
scaler = StandardScaler()
X_scaled_usage = scaler.fit_transform(X_usage)

# Choose the optimal K for usage pattern (e.g., based on the elbow method)
distortions = []
K_range = range(1, 11)
for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=0)
    kmeans.fit(X_scaled_usage)
    distortions.append(kmeans.inertia_)
```

```python
# Plot the elbow graph for usage pattern
plt.figure(figsize=(8, 4))
plt.plot(K_range, distortions, marker='o')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Distortion')
plt.title('Elbow Method for Optimal K (Usage Pattern)')
plt.show()

# Based on the elbow method, choose the optimal K for usage pattern
optimal_k_usage = 3  # Adjust this based on your analysis

# Perform K-means clustering for usage pattern
kmeans_usage = KMeans(n_clusters=optimal_k_usage, random_state=0)
clusters_usage = kmeans_usage.fit_predict(X_scaled_usage)

# Add cluster labels for usage pattern to the original dataset
data['Cluster_Usage'] = clusters_usage
```

```python
data
```

```python
plt.figure(figsize=(8, 6))
for cluster in range(optimal_k_usage):
    plt.scatter(
        data[data['Cluster_Usage'] == cluster]['2 W'],
        data[data['Cluster_Usage'] == cluster]['3 W'],
        label=f'Cluster {cluster + 1}',
        alpha=0.7
    )

plt.xlabel('2 W')
plt.ylabel('3 W')
plt.title('Scatter Plot of Usage Pattern Clusters')
plt.legend()
plt.grid()
plt.show()
```

## Usage Pattern Clustering:

Cluster Identification: Clustering the data based on usage patterns (e.g., '2 W' and '3 W') has revealed distinct clusters of entities. These clusters likely represent different types of users or entities in the market.

Market Segmentation: The clusters can be seen as market segments with unique usage behaviors. Understanding these segments can help in tailoring marketing strategies and product offerings to specific customer groups.

Targeted Marketing: By identifying high-usage clusters, you can target marketing efforts and promotions to those segments that are more likely to engage with your products or services.

Based on analysis of the centroid values and the distribution of data points within each cluster, you can make observations about the usage patterns. For example:

Cluster 1 represents low usage in both '2 W' and '3 W.'
Cluster 2 represents moderate usage in both '2 W' and '3 W.'
Cluster 3 represents high usage in both '2 W' and '3 W.'

```python


# Extract relevant columns for temporal trend
X_temporal = data[['TOTAL']]

# Convert 'YEAR' to timestamp
X_temporal['YEAR'] = pd.to_datetime(data['YEAR']).astype('int64') // 10**9

# Standardize the data
scaler = StandardScaler()
X_scaled_temporal = scaler.fit_transform(X_temporal)

# Choose the optimal K for temporal trend (e.g., based on the elbow method)
distortions = []
K_range = range(1, 11)
for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=0)
    kmeans.fit(X_scaled_temporal)
    distortions.append(kmeans.inertia_)


```

```python
# Plot the elbow graph for temporal trend
plt.figure(figsize=(8, 4))
plt.plot(K_range, distortions, marker='o')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Distortion')
plt.title('Elbow Method for Optimal K (Temporal Trend)')
plt.show()

# Based on the elbow method, choose the optimal K for temporal trend
optimal_k_temporal = 3  # Adjust this based on your analysis

# Perform K-means clustering for temporal trend
kmeans_temporal = KMeans(n_clusters=optimal_k_temporal, random_state=0)
clusters_temporal = kmeans_temporal.fit_predict(X_scaled_temporal)

# Add cluster labels for temporal trend to the original dataset
data['Cluster_Temporal'] = clusters_temporal
```

```python
plt.figure(figsize=(10, 6))
for cluster in range(optimal_k_temporal):
    cluster_data = data[data['Cluster_Temporal'] == cluster]
    plt.plot(cluster_data['YEAR'], cluster_data['TOTAL'], label=f'Cluster {cluster + 1}')

plt.xlabel('Year')
plt.ylabel('TOTAL')
plt.title('Temporal Trends by Cluster')
plt.legend()
plt.grid()
plt.show()
```

## Temporal Trend Clustering:

Cluster Interpretation: Clustering based on temporal trends (e.g., 'TOTAL' over time) has grouped entities with similar historical usage patterns. Different clusters may represent entities with different growth trajectories or seasonality.

Growth Opportunities: Identify clusters that exhibit consistent growth over time. These clusters may represent market segments with growth potential, suggesting areas where you can focus resources.

Declining Trends: Clusters with declining trends may indicate markets or customer groups that need attention or new strategies to reverse the decline.

For example, you might find that Cluster 1 represents entities with steady growth, Cluster 2 exhibits seasonal fluctuations, and Cluster 3 shows a declining trend in 'TOTAL' values.
